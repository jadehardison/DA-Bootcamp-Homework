# Observable trends based on the data include:

Out of the 780 purchases completed by players, roughly 84% were male, 14% were female, and a small percentage of players did not disclose gender.

In regards to age demographics the majority of players are between the ages of 20-25 at a whopping 40%.  This age group generates the most profit followed by the 15-19 age group at 26%. 

Males generate far more profit on average than female players. 

Oathbreaker, Last Hope of the Breaking Storm was the most popular game.

```python
import pandas as pd
import numpy as np
```

```python
#pymoli_df = pd.read_csv('C:\\Users\\jadeh\\Downloads\\purchase_data.csv')
#display(pymoli_df)
#print(len(pymoli_df))

pymoli_df = pd.read_csv("Resources/purchase_data.csv")
```

```python
total_players = len(pymoli_df["SN"].value_counts())

player_count = pd.DataFrame({"Total Players":[total_players]})
player_count

```

```python
unique_items = len((pymoli_df["Item ID"]).unique())
average_price = (pymoli_df["Price"]).mean()
num_purchases = (pymoli_df["Purchase ID"]).count()
total_rev = (pymoli_df["Price"]).sum()

summary_df = (pd.DataFrame({"Number of Unique Items":[unique_items],
                      "Average Price":[average_price],
                      "Number of Purchases":[num_purchases],
                       "Total Revenue":[total_rev]}))

display(summary_df)
print(len(summary_df))
```

```python
gender_demo = pymoli_df.groupby("Gender")
totalgender_count = (gender_demo.nunique()["SN"])
totalgender_count.count()
percentage_of_genders = totalgender_count / total_players * 100
gender_df = pd.DataFrame({"Total Count": totalgender_count,
                         "Percentage of Players": percentage_of_genders,})

gender_df.index.name = None
gender_df

```

```python

purchasegen_count = gender_demo["Purchase ID"].count()
purchasegen_avg = gender_demo["Price"].mean()
purchasegen_avg_total = gender_demo["Price"].sum()
purchasegen_per_person = purchasegen_avg_total/totalgender_count

gender_analysis = pd.DataFrame({"Purchase Count": purchasegen_count,
                               "Average Purchase Price": purchasegen_avg,
                               "Total Purchase Value": purchasegen_avg_total,
                               "Avg Total Purchase per Person": purchasegen_per_person})
gender_analysis


```

```python
age_bins = [0, 9.50, 15.50, 20.50, 25.50, 30.50, 35.50, 40.50, 100.50] 
group_ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

pymoli_df["Age Group"] = pd.cut(pymoli_df["Age"], age_bins, labels=group_ages)
pymoli_df

grouped_by_age = pymoli_df.groupby("Age Group")

totalage_count = grouped_by_age["SN"].nunique()

age_percentage = (totalage_count/total_players) * 100
age_demo = pd.DataFrame({"Total Count": totalage_count, "Percentage of Players": age_percentage})

age_demo.index.name = None

age_demo
```

```python
age_purchase_count = grouped_by_age["Purchase ID"].count()

age_avg_purchase_price = grouped_by_age["Price"].mean()

age_total_purchase_value = grouped_by_age["Price"].sum()

age_avg_purchase_per_person = age_total_purchase_value/totalage_count

age_analysis_demo = pd.DataFrame({"Purchase Count": age_purchase_count,
                                 "Average Purchase Price": age_avg_purchase_price,
                                 "Total Purchase Value": age_total_purchase_value,
                                 "Average Purchase Total per Person": age_avg_purchase_per_person})

age_analysis_demo.index.name = None
age_analysis_demo
```

```python
top_spenders = pymoli_df.groupby("SN")
top_purchase_count = top_spenders["Purchase ID"].count()
top_avg_price = top_spenders["Price"].mean()
top_purchases = top_spenders["Price"].sum()

top_spenders_df = pd.DataFrame({"Purchase Count": top_purchase_count,
                                "Average Purchase Price": top_avg_price,
                                "Total Purchase Value": top_purchases})

formatted_purchasevalue = top_spenders_df.sort_values(["Total Purchase Value"], ascending=False).head()
formatted_purchasevalue
```

```python
items = pymoli_df[["Item ID", "Item Name", "Price"]]

item_stats = items.groupby(["Item ID","Item Name"])

purchasecount_item = item_stats["Price"].count()

purchase_value = (item_stats["Price"].sum()) 

item_price = purchase_value/purchase_count_item

most_popular_items = pd.DataFrame({"Purchase Count": purchasecount_item, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":purchase_value})

mostpopular_formatted = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()
mostpopular_formatted
```
